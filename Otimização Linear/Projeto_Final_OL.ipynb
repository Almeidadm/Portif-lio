Primeiramente, imaginemos um cliente que quer maximizar o seu lucro, sem se importar com o risco de cada ação. Entretanto, para evitar grandes perdas, quer comprar no mínimo 5 ações de cada empresa da bolsa e também não quer alocar mais do que 25% de seu capital em um dos setores. O seu investimento será um aporte de R$ 20.000 e o deixará parado por 5 anos.


```python
#Projeto

prob = pulp.LpProblem('Escolha de investimentos', pulp.LpMaximize)

#variaveis

x1 = pulp.LpVariable('Transmissao Paulista (TRPL4)', lowBound = 5, cat = 'Integer')
x2 = pulp.LpVariable('Engie (EGIE3)', lowBound = 5, cat = 'Integer')
x3 = pulp.LpVariable('Taesa (TAEE3)', lowBound = 5, cat = 'Integer')
x4 = pulp.LpVariable('Itau (ITUB3)', lowBound = 5, cat = 'Integer')
x5 = pulp.LpVariable('Banco do Brasil (BBAS3)', lowBound = 5, cat = 'Integer')
x6 = pulp.LpVariable('Bradesco (BBDC3)', lowBound = 5, cat = 'Integer')
x7 = pulp.LpVariable('BCFF11', lowBound = 5, cat = 'Integer')
x8 = pulp.LpVariable('IRDM11', lowBound = 5, cat = 'Integer')
x9 = pulp.LpVariable('HCTR11', lowBound = 5, cat = 'Integer')
x10 = pulp.LpVariable('Hypera farma (HYPE3)', lowBound = 5, cat = 'Integer')
x11 = pulp.LpVariable('Fleury (FLRY3)', lowBound = 5, cat = 'Integer')
x12 = pulp.LpVariable('OdontoPrev (ODPV3)', lowBound = 5, cat = 'Integer')
x13 = pulp.LpVariable('Embraer (EMBR3)', lowBound = 5, cat = 'Integer')
x14 = pulp.LpVariable('Gol (GOLL4)', lowBound = 5, cat = 'Integer')
x15 = pulp.LpVariable('Weg (WEG3)', lowBound = 5, cat = 'Integer')

#funcao objetivo

FO = 24.55*x1*((1+0.2)**5)+38.1*x2*((1+0.1)**5)+13.33*x3*((1+0.25)**5)+28.2*x4*((1+0.2)**5)+31.86*x5*((1+0.13)**5)+20.44*x6*((1+0.18)**5)+75.14*x7*((1+0.08)**5)+117.29*x8*((1+0.13)**5)+123.1*x9*((1+0.19)**5)+36.75*x10*((1+0.12)**5)+23.09*x11*((1+0.09)**5)+13.55*x12*((1+0.1)**5)+19.01*x13*((1+0.09)**5)+19.85*x14*((1+0.1)**5)+36.34*x15*((1+0.1)**5)
prob += FO

#Restrições

MaxRenda = 24.55*x1+38.1*x2+13.33*x3+28.2*x4+31.86*x5+20.44*x6+75.14*x7+117.29*x8+123.1*x9+36.75*x10+23.09*x11+13.55*x12+19.01*x13+19.85*x14+36.34*x15
prob += (MaxRenda <= 20000)

QntEnergia = 24.55*x1+38.1*x2+13.33*x3
prob += (QntEnergia <= 5000)

QntBanco = 28.2*x4+31.86*x5+20.44*x6
prob += (QntBanco <= 5000)

QntFI = 75.14*x7+117.29*x8+123.1*x9
prob += (QntFI <= 5000)

QntSaude = 36.75*x10+23.09*x11+13.55*x12
prob += (QntSaude <= 5000)

QntIndus = 19.01*x13+19.85*x14+36.34*x15
prob += (QntIndus <= 5000)

#Resultado
print (prob)

optimization_result = prob.solve()

assert optimization_result == pulp.LpStatusOptimal

print('Serão comprados:')
for var in prob.variables():
  print('{} acoes da empresa {}'.format(var.value(), var.name))
print('O lucro final será de: R${:0.2f}' .format(FO.value()))



```

Quando falamos de investimentos, quanto maiores os riscos mais os investimentos devem retornar para se compensar. Entretanto nos testes 2 e 3 pudemos perceber que quando maximizamos os riscos o retorno dos investimentos foram menores, o que nao faz sentido, pois porque iríamos assumir mais risco sendo que investimentos com menos riscos trazem retornos maiores? Logo, podemos concluir que talvez as análises que utilizamos não foram as mais eficientes e corretas. O que nos indica que essas análises mais simples e lineares nao sao as melhores análises, até porque se fossem todos ganhariam facil dinheiro na bolsa de valores.
